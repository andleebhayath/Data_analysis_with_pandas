# Adding a title to FacetGrid
# and adjusting hieght of suptitle
# setting axis labels

```python
g=sns.catplot(x='sex',y='tip',data=df,kind='bar',ci=None)
g.fig.suptitle('Sex',y=1.07)#y  mean adjusting height of subtitle 
g.set(xlabel='sex',ylabel='tip')
plt.show()
```

# Rotating x_axis ticks labels

```python
g=sns.catplot(x='sex',y='tip',data=df,kind='bar',ci=None)
g.fig.suptitle('Sex',y=1.07)#y  mean adjusting height of subtitle 
g.set(xlabel='sex',ylabel='tip')
plt.xticks(rotation=90)
plt.show()
```
